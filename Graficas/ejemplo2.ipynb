```python
from mpl_toolkits.axes_grid1 import host_subplot
from mpl_toolkits import axisartist
import matplotlib.pyplot as plt
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
datos = pd.read_csv('./continous_data.csv')

NO2= datos['NO2 (ug/m3)']

NO= datos['NO (ug/m3)']

CO= datos['CO (ug/m3)']

O3= datos['O3 (ug/m3)']

Uptime= datos['Uptime (s)'] 



```

```python

from calendar import c
fig, ax = plt.subplots()


ax.plot(Uptime,NO2,label='NO2',lw=0.8, alpha=0.8)
ax.set(xlabel='Tiempo')
ax.set(ylabel='(ug/m3)')

ax1=ax.twinx()
ax1.plot(Uptime,CO,label='CO',c='orange',lw=0.8, alpha=0.8)
ax1.set(ylabel='(ug/m3)')
ax1.tick_params(axis='y', colors='orange')
ax1.legend(loc=2)


ax2=ax.twinx()
ax2.spines["right"].set_position(('axes',1.2))
ax2.plot(Uptime,NO,label='NO',c='green',lw=0.8, alpha=1)
ax2.set(ylabel='(ug/m3)')
ax2.tick_params(axis='y', colors='green')
ax2.legend(loc=5)


ax.plot(Uptime,O3,label='O3', c='red',lw=0.7, alpha=0.5)

ax.grid()
ax.legend()
fig.savefig('Medidas 20-23 Mayo.png',dpi=600)
plt.show()



```
